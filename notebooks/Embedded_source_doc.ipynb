## Source 1 
### Question 1 

This document compares different indoor localization methods.

The first category concerns the use of Wi‑Fi. This technology is particularly attractive because most buildings are already equipped with Wi‑Fi infrastructure. The document presents several Wi‑Fi-based techniques. A first method is RSS ranging or fingerprinting, which uses the received signal strength to estimate a distance or position. A second method is CSI fingerprinting, which estimates the position by comparing the characteristics of received signals with a reference database. A third method is AoA (Angle of Arrival), which can determine a position using a single Wi‑Fi access point. However, this approach requires specialized antennas capable of receiving and analyzing the signals. Finally, the FTM/RTT (Fine Timing Measurement / Round Trip Time) method measures the signal travel time to estimate the distance between the transmitter and the receiver.

The second category is based on the use of UWB (Ultra-Wide Band). This technology is particularly well suited for indoor localization thanks to its large bandwidth and low power consumption compared with other solutions. As with Wi‑Fi, several methods are available. RTT ranging and AoA rely on the same principles as their Wi‑Fi counterparts.

The third category concerns BLE (Bluetooth Low Energy). Today, BLE is one of the most widely used technologies for indoor localization, mainly because it is directly integrated into most smartphones. This approach relies on the analysis of signals transmitted by Bluetooth beacons to estimate the user’s position.

The fourth category involves the use of IMUs (Inertial Measurement Units). IMUs make it possible to estimate the relative motion of a smartphone in order to track a user’s movement. To do so, they exploit the various sensors embedded in the device. The accelerometer detects steps from variations in the measured signal. The gyroscope determines the direction of movement by measuring angular velocity. Finally, the magnetometer provides a reference direction using the Earth's magnetic field.


### Question 2 

Several assumptions were made when comparing different indoor localization methods.

The assumptions related to localization methods state that the user's position corresponds to the nearest reference point. They also assume that the position can be estimated through trilateration using three reference points. Finally, they assume that each reference point is associated with a unique fingerprint.

The assumptions concerning sensors are based on the use of Pedestrian Dead Reckoning (PDR) and the Kalman filter for localization. They also take into account the instability of the Received Signal Strength (RSS) as well as the non-uniform attenuation of BLE signals.

The assumptions related to the environment consider that signals propagate directly between the transmitter and the receiver, without significant effects from reflection, refraction, or diffraction. They also assume the presence of a smartphone and reference beacons to enable localization. Finally, they assume that GPS is unavailable in indoor environments.

In addition, BLE was preferred over Wi‑Fi because the BLE beacons are already installed, easily identifiable, and their positions are known. In contrast, Wi‑Fi access points are not necessarily documented or under our control, making them more difficult to use for localization purposes.

## Source 2 
### Question 1

This document presents an approach for determining the path followed by a user inside a building. The approach is based on the combination of a particle filter and Pedestrian Dead Reckoning (PDR).

The first step consists of initializing the particles of the filter. Each particle represents a possible hypothesis of the user's position. These particles are then propagated using the motion model provided by PDR and weighted according to BLE observations.

The second step involves applying PDR, a technique that estimates the user's current position by adding a displacement vector derived from inertial sensor measurements.

The third step consists of comparing the estimated positions with the BLE measurements in order to evaluate the consistency of each particle. The particle weights are then updated according to this consistency.

Finally, a resampling step retains the most likely particles while eliminating the least consistent hypotheses. This process progressively improves the estimation of the trajectory followed by the user.


### Question 2

In its implementation, the document makes several assumptions to evaluate the performance of the proposed approach.

First, it assumes that the smartphone, held vertically, is aligned with the user's walking direction. This assumption simplifies trajectory estimation from inertial sensor data.

The study also assumes that all BLE beacons transmit at the same power level and that proximity estimation relies solely on the received signal strength (RSS). Therefore, potential variations in transmission power between beacons are not taken into account.

Finally, the document considers that measurements obtained from BLE, IMUs, and the particle filter are affected by various external sources of error, such as measurement noise, radio interference, and sensor inaccuracies. The proposed approach must therefore be robust enough to handle these imperfections.


### Question 3

This document was selected because it closely matches our project context. Indeed, the objective is to find a technique capable of reconstructing a user's trajectory inside a building, which is precisely the problem addressed in this article.

Furthermore, one of the constraints of our project was to use an approach combining BLE and IMU technologies together with a particle filter. The method proposed in this document satisfies these requirements, making it a particularly relevant reference for the development of our solution.


### Question 4

The proposed approach is very similar to the one we implement: it reconstructs a user's trajectory inside a building by combining BLE, IMU, and a particle filter, and its filter loop (initialize the particles, propagate them with the PDR motion model, weight them with BLE observations, and resample) is essentially the one we use.

There are, however, important differences, and in several respects our work extends theirs. First, Jin et al. weight the particles using a range-free weighted-centroid BLE position, whereas we weight each particle directly with a log-distance path-loss likelihood on the observed RSSI. Second, their heading is obtained from a complementary filter that fuses the magnetometer and gyroscope with the phone held upright and aligned with the walking direction; because our phone is carried in a trouser pocket at an unknown orientation, and because indoor magnetic fields are unreliable, we estimate heading from the gyroscope alone (projected onto gravity and bias-corrected).

Most importantly, Jin et al. explicitly list as future work the extension of their method "to the situation when there are map constraints, and the placement of the smartphone is unconstrained" — which is precisely what our project does: we incorporate building/corridor constraints, handle floor transitions through staircases across two floors, and use a pocket-placed phone. Our approach can therefore be seen as taking their fused BLE + PDR particle filter in exactly the directions they identified as open problems.


### Question 5

This document helped us select this method for our project. Indeed, the proposed approach takes into account both the constraints and the objectives of the project. Furthermore, the results presented in the article demonstrate the effectiveness of this method in a context similar to ours. These performances therefore reinforce the relevance of our choice and provide strong support for its implementation.

## Source 3 
### Question 1

This document compares three indoor localization algorithms based on IMU and ToA (Time of Arrival) measurements.

The first algorithm is MAP (Maximum a Posteriori), which aims to maximize the posterior probability in order to estimate the user's position. ToA measurements provide an estimate of the distance to radio anchors. In a two-dimensional environment, each measurement theoretically constrains the user's position to a circle centered on the corresponding anchor.

The second algorithm is the Extended Kalman Filter (EKF), which uses a dynamic model together with the available measurements to estimate the position while accounting for sensor uncertainties.

The third algorithm is the particle filter, which represents the position using a set of particles corresponding to different hypotheses. These particles are continuously updated and weighted using the available measurements to obtain the most probable estimate of the user's position.


### Question 2

The document makes several assumptions regarding the modeling of the problem. It assumes that the user's motion (direction, speed, etc.) contains a random component. It is also assumed that the Extended Kalman Filter (EKF) relies on a Gaussian noise model, whereas the particle filter is based on a Markov process. In this approach, the current state is estimated from the previous state, the motion model, and the available observations. The particle filter therefore relies on a sequential probabilistic model of position evolution.

Assumptions regarding the environment and sensors are also considered. Localization is studied in a context where GPS is unavailable, which is typical of indoor environments. Finally, it is assumed that Time of Arrival (ToA) measurements provide ideal distance estimates, meaning that the user's position theoretically lies on the circles defined around the different radio anchors.


### Question 3

This source was selected to support our choice between the Extended Kalman Filter (EKF) and the particle filter. Indeed, the document allows a comparison of the performance of these two localization methods in a context similar to ours, based on the fusion of inertial sensor data and radio measurements.

The study therefore provides concrete insights into the strengths and limitations of each approach. It helps us determine which method is best suited to our constraints and to the characteristics of our indoor localization environment.


### Question 4

Regarding the similarities between the approach presented in this document and our project, several elements are shared. In both cases, position estimation relies partly on data provided by IMU sensors. In addition, both Kalman filters and particle filters are considered as potential solutions for data fusion and trajectory estimation.

As for the differences, the document also relies on Time of Arrival (ToA) measurements for localization, whereas our project uses BLE measurements. Furthermore, the study analyzes the performance of the Maximum a Posteriori (MAP) algorithm, which is not part of the methods considered in our project. Therefore, although the overall context is similar, some of the technologies and algorithms studied differ from those adopted in our approach.


### Question 5

In light of the results presented in this document, our choice of using a particle filter was reinforced. Indeed, in the experimental configuration studied, the particle filter achieved better localization accuracy than the Extended Kalman Filter (EKF).

However, the study also showed that the best performance was obtained with the Maximum a Posteriori (MAP) algorithm. Despite these results, this solution was not selected for our project due to several constraints, particularly regarding its implementation and its suitability for the project's data and objectives.

Therefore, the conclusions of this article support the relevance of our choice to use a particle filter while highlighting the existence of potentially more accurate approaches under different application conditions.
